Exercise 1 — Input + Math + Strings

Start with the initial comment: to get personal information from the user

Ask for name, age, height (cm), and weight (kg)

Print lowercase, uppercase, and the length of the name

Convert height to meters and print BMI = weight / height²

Print weight divided by π

Print age in days

Split the name into characters and print them one per line

```python
import math
name = input("Enter your name ")
print(name.lower())
print(name.upper())
print(len(name))

age = int(input("Enter your age "))
height = int(input("Enter your height in cm "))
weight = int(input("Enter your weight in kg "))

height = height/100
print("Height in meters : ", height)
print("BMI", weight//(height**2) )

print("Weight divided by pi ", weight//math.pi)

print("Age in days : ", age*365)

print("Characters in name : ")
for char in name:
    print(char)
```

Exercise 3 — List Processing

Prompt the user to enter 10 integers and store them in a list, then print:

Mean (average)

Min and Max

Count of numbers > 50

All odd numbers

All numbers divisible by 3

The list sorted (ascending and descending)

The list reversed

```python
input_list = input("Enter 10 integers separated by space ").split()
input_list = [int(i) for i in input_list]
mean = sum(input_list)/len(input_list)
print("Mean :", mean)

print("Min :" , min(input_list))
print("Max :" , max(input_list))
greater_than_50 = [i for i in input_list if i > 50]
print("Count of numbers > 50 :", len(greater_than_50))

odd_numbers = [i for i in input_list if i%2 != 0]
print("All odd numbers :", odd_numbers)

divisible_by_3 = [i for i in input_list if i%3 == 0]
print("All numbers divisible by 3 :", divisible_by_3)

input_list.sort()
print("Sorted list (ascending) :", input_list)

input_list.sort(reverse=True)
print("Sorted list (descending) :", input_list)

input_list.reverse()
print("Reversed list :", input_list)

```

Create 3 functions:

def add(a, b): return a + b

def multiply(a, b): return a * b

def power(a, b): return a ** b

Then write a main program that:

Exercise 5 — Dictionary Storage

Create a program that asks for:

5 student names

1 score for each student

Store them in a dictionary like:

            {"Alice": 88, "Bob": 92, ...}

Then print:

Class average
Student with highest score
All students who scored above 90

```python
names = input("Enter 5 student names separated by space ").split()
scores = input("Enter 5 scores separated by space ").split()
scores = [int(i) for i in scores]
scores_dict = dict(zip(names, scores))
print(scores_dict)

average = sum(scores)/len(scores)
print("Class average :", average)

highest_score = max(scores_dict, key=scores_dict.get)
print("Student with highest score :", highest_score)

above_90 = [name for name, score in scores_dict.items() if score > 90]
print("Students who scored above 90 :", above_90)
```
