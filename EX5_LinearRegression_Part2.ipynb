For a health insurance company to make money, it needs to collect more in yearly premiums than it spends on medical care to its beneﬁciaries.

As a result, insurers invest a great deal of time and money in developing models that accurately forecast medical expenses for the insured population.

Medical expenses are diﬃcult to estimate because the costliest conditions are rare and seemingly random.

Still, some conditions are more prevalent for certain segments of the population. For instance, lung cancer is more likely among smokers than non-smokers, and heart disease may be more likely among the obese.

The goal of this analysis is to use patient data to estimate the average medical care expenses for such population segments.

These estimates can be used to create actuarial tables that set the price of yearly premiums higher or lower, depending on the expected treatment costs.

## Train OLS Model (statsmodels)

Fit a linear regression model using statsmodels.api.OLS() with all available features.


```python
import statsmodels.api as sm

# Define the dependent variable (y)
y = df_encoded['charges']

# Define the independent variables (X) by dropping 'charges'
X = df_encoded.drop('charges', axis=1)

# Add a constant (intercept) to the independent variables
X = sm.add_constant(X)

# Instantiate and fit the OLS model
model = sm.OLS(y, X).fit()

print("OLS model fitted successfully.")
```

### OLS Model Performance Evaluation

The model.summary() output provides a comprehensive overview of the OLS regression model's performance and statistical properties:

**1. R-squared and Adjusted R-squared:**

*   **R-squared (0.751):** This indicates that approximately 75.1% of the variance in medical charges (charges) can be explained by the independent variables included in this model. This is a relatively high value for a social science or real-world dataset, suggesting that the model has a good explanatory power.

It accounts for the number of predictors in the model, providing a more honest estimate of the model's explanatory power.

**2. F-statistic and Prob (F-statistic):**

*   **F-statistic (500.8):** This is a test for the overall significance of the regression model. It tests the null hypothesis that all the regression coefficients are equal to zero (i.e., that none of the independent variables have a significant linear relationship with the dependent variable).
*   **Prob (F-statistic) (0.00):** The p-value associated with the F-statistic is extremely small (effectively zero). This means we can reject the null hypothesis, concluding that the model as a whole is statistically significant. At least one of the independent variables is useful in predicting medical charges.

**3. Coefficients (coef) and their p-values (P>|t|) for Individual Predictors:**

*   **Statistically Significant Features (P>|t| < 0.05):**
    *   age (P>|t| = 0.000): Highly significant. As age increases, charges significantly increase.
    *   bmi (P>|t| = 0.000): Highly significant. As BMI increases, charges significantly increase.
    *   children (P>|t| = 0.001): Significant. More children are associated with significantly higher charges.
    *   smoker_no (P>|t| = 0.000) and smoker_yes (P>|t| = 0.000): Both highly significant. This indicates that being a smoker or non-smoker has a profound and statistically significant impact on medical charges.

*   **Not Statistically Significant Features (P>|t| > 0.05):**
    *   const (P>|t| = 0.491): The intercept is not statistically significant at conventional levels.
    *   sex_female (P>|t| = 0.759) and sex_male (P>|t| = 0.437): Neither gender dummy variable is individually significant. This means that, after accounting for other factors, there isn't a statistically significant difference in charges between male and female in the way they are represented here.
    *   region_northeast (P>|t| = 0.088): Borderline significant; might be considered significant at a higher alpha level.
    *   region_northwest (P>|t| = 0.596), region_southeast (P>|t| = 0.115), region_southwest (P>|t| = 0.151): These region dummy variables are not statistically significant. This suggests that, after controlling for other variables, the differences in charges across regions are not statistically robust.

**4. Diagnostic Statistics:**

*   **Omnibus (300.366) and Prob(Omnibus) (0.000):** This is a test of the normality of the residuals. A low p-value (0.000) suggests that the residuals are not normally distributed. This aligns with our earlier observation that charges itself is not normally distributed and is right-skewed, which often leads to non-normal residuals in linear regression.
*   **Jarque-Bera (JB) (718.887) and Prob(JB) (7.86e-157):** Similar to the Omnibus test, this also tests for normality of residuals. The extremely small p-value further confirms that the residuals are not normally distributed, exhibiting significant skew (Skew: 1.211) and high kurtosis (Kurtosis: 5.651).
*   **Skew (1.211):** Positive skewness indicates a longer tail on the right side of the residual distribution, consistent with the right-skewness of the dependent variable.
*   **Kurtosis (5.651):** High kurtosis suggests that the residual distribution has heavier tails and a sharper peak than a normal distribution.
*   **Durbin-Watson (2.088):** This statistic tests for the presence of autocorrelation in the residuals. A value close to 2 (which 2.088 is) suggests no significant autocorrelation. This is good, as autocorrelated residuals violate an assumption of OLS.
*   **Cond. No. (5.36e+17):** The Condition Number is extremely large. This indicates severe **multicollinearity**, meaning that some independent variables are highly correlated with each other. This is a critical issue because it makes the coefficient estimates unstable and difficult to interpret. The note [2] The smallest eigenvalue is 1.22e-29. This might indicate that there are strong multicollinearity problems or that the design matrix is singular. explicitly highlights this. The dummy variables for sex, smoker, and region are perfectly multicollinear if all categories are included along with an intercept, which is why statsmodels automatically drops one category or uses a different approach (like setting one category as the baseline by effectively absorbing its effect into the intercept). However, the remaining sex_female, sex_male as well as smoker_no, smoker_yes will still cause multicollinearity as they sum to 1. One dummy variable for each categorical feature should be dropped to avoid perfect multicollinearity, or one-hot encoding without dropping the first category could lead to this issue when an intercept is present.

**5. How the model is performing? (Summary)**

The OLS model appears to be **performing reasonably well in terms of explaining variance** in medical charges, with an R-squared of 0.751. The model is also **overall statistically significant**. Key factors like age, bmi, and smoker status (as indicated by smoker_no and smoker_yes) are strong and significant predictors of charges.

However, the model has **significant diagnostic issues**:

*   **Non-normal residuals:** The Omnibus and Jarque-Bera tests clearly indicate that the residuals are not normally distributed, which violates one of the core assumptions of OLS regression. This could affect the validity of hypothesis tests (p-values) and confidence intervals.
*   **Severe Multicollinearity:** The extremely high Condition Number points to significant multicollinearity among the predictors. This can inflate the standard errors of the coefficients, making some predictors appear statistically insignificant when they might be, and makes it difficult to ascertain the individual impact of correlated predictors. This is likely due to the dummy encoding where all categories of a categorical variable were included alongside the intercept, which creates perfect multicollinearity (e.g., sex_female + sex_male always equals 1).

**Overall, while the model has high explanatory power, the violations of OLS assumptions (non-normal residuals and severe multicollinearity) suggest that the current model's coefficient interpretations and statistical inferences should be approached with caution. Transformations of the dependent variable (e.g., log-transforming charges) or addressing multicollinearity by dropping one dummy variable per categorical feature are potential next steps to improve the model's reliability and validity.**

## Implement SGDRegressor for Mini-batch Gradient Descent

Use SGDRegressor function, from sklearn.linear_model, to implement a simulation of the Mini-batch Gradient Descent.

```python
import numpy as np
import random

# Initialize the SGDRegressor
# Set max_iter to a large number so it doesn't stop early if convergence is slow
# Set shuffle=True to shuffle data each epoch internally (though we also shuffle manually below)
# Set warm_start=True to reuse parameters from previous partial_fit calls
sgd_regressor_minibatch = SGDRegressor(max_iter=1000, tol=1e-3, learning_rate='invscaling', eta0=0.01, random_state=42, shuffle=True, warm_start=True)

# Define mini-batch parameters
batch_size = 32  # Typical mini-batch size
n_epochs = 20
m = X_train_scaled.shape[0] # Total number of training samples (using scaled X_train)

# Manually iterate over epochs and mini-batches
for epoch in range(n_epochs):
    # Shuffle data indices at the start of each epoch
    indices = np.random.permutation(m)
    X_shuffled = X_train_scaled.iloc[indices] # Use iloc for DataFrame indexing
    y_shuffled = y_train.iloc[indices] # Use iloc for Series indexing

    for i in range(0, m, batch_size):
        # Extract mini-batch
        X_batch = X_shuffled[i:i + batch_size]
        y_batch = y_shuffled[i:i + batch_size]

        # Perform incremental learning on the current mini-batch
        sgd_regressor_minibatch.partial_fit(X_batch, y_batch)
```

## Key Hyperparameters of Each Implementation

Let's review the key hyperparameters for each of the linear regression implementations used:

### 1. sklearn.linear_model.LinearRegression (Normal Equation)

This implementation typically solves the linear regression problem using the Ordinary Least Squares (OLS) method, often through a closed-form solution (Normal Equation) or via singular value decomposition (SVD). For LinearRegression in scikit-learn, there are relatively few hyperparameters related to the optimization algorithm itself because it's a direct solution method rather than an iterative one.

**Key Hyperparameters:**
*   fit_intercept: (default=True) Boolean, whether to calculate the intercept for this model. If set to False, no intercept will be used in calculations (e.g., data is expected to be already centered).
*   copy_X: (default=True) Boolean, if True, X will be copied; else, it may be overwritten. Ensures that the original input X is not modified.
*   n_jobs: (default=None) The number of jobs to use for the computation. -1 means using all processors. This is for parallelizing calculations, especially for multiple targets.
*   positive: (default=False) Boolean, when set to True, forces the coefficients to be positive. (Added in scikit-learn 0.24).

### 2. sklearn.linear_model.SGDRegressor (Stochastic Gradient Descent)

SGDRegressor is a linear model fitted by minimizing a regularized empirical loss with Stochastic Gradient Descent (SGD). In pure Stochastic Gradient Descent, only one sample is used per iteration. The default batch_size of 1 (implicitly in how SGDRegressor works on a sample-by-sample basis during each epoch by default, though it processes n_samples in partial_fit or a single fit) can be considered as such.

**Key Hyperparameters:**
*   loss: (default='squared_error') The loss function to be used. For linear regression, squared_error is the standard choice. Other options include huber, epsilon_insensitive, squared_epsilon_insensitive for robust regression.
*   penalty: (default='l2') The penalty (regularization term) to be used. Options are l2 (Ridge), l1 (Lasso), or elasticnet (combination of L1 and L2). No penalty can be specified with None.
*   alpha: (default=0.0001) Constant that multiplies the regularization term. Higher values mean stronger regularization.
*   l1_ratio: (default=0.15) The Elastic Net mixing parameter, with 0 <= l1_ratio <= 1. Only used if penalty is 'elasticnet'.
*   fit_intercept: (default=True) Same as in LinearRegression.
*   max_iter: (default=1000) The maximum number of passes over the training data (epochs).
*   tol: (default=1e-3) The stopping criterion. The training will stop when (loss > best_loss - tol) for n_iter_no_change consecutive epochs.
*   shuffle: (default=True) Whether to shuffle the training data after each epoch.
*   eta0: (default=0.01) The initial learning rate for the 'constant', 'invscaling', or 'adaptive' schedules.
*   learning_rate: (default='invscaling') The learning rate schedule. Options include constant, optimal, invscaling, adaptive.
*   power_t: (default=0.25) The exponent for inverse scaling learning rate. Used when learning_rate is 'invscaling'.
*   early_stopping: (default=False) Whether to use early stopping to terminate training when validation score is not improving.
*   validation_fraction: (default=0.1) The proportion of training data to set aside as validation set for early stopping.
*   n_iter_no_change: (default=5) Number of iterations with no improvement to wait before stopping.
*   random_state: (default=None) Used for shuffling the data. Useful for reproducibility.

### 3. sklearn.linear_model.SGDRegressor (Mini-batch Gradient Descent Simulation)

This is essentially the same SGDRegressor as above, but with specific choices for how batch_size is handled. In scikit-learn's SGDRegressor, the batch_size parameter itself is not directly exposed as it is in some other frameworks (like Keras). Instead, SGDRegressor inherently processes samples one by one (stochastic) or in small batches depending on internal implementation details and how the fit method iterates. To simulate mini-batch, one would typically manage the batches manually or rely on its internal mechanisms to process data efficiently. However, the hyperparameters controlling the learning process are the same.

**Key Hyperparameters (same as for Stochastic GD, emphasizing those relevant for batch behavior):**
*   loss: (default='squared_error') Chosen for linear regression.
*   max_iter: (default=1000) Maximum epochs.
*   tol: (default=1e-3) Stopping criterion.
*   eta0: (default=0.01) Initial learning rate.
*   learning_rate: (default='invscaling') Determines how the learning rate changes over iterations.
*   early_stopping: (default=False, but set to True in our code) Controls termination based on validation performance.
*   n_iter_no_change: (default=5, used in our code) Number of iterations for early stopping.

## Evaluate Regression Model Performance Metrics

Find different metrics to evaluate your regression model performance.

### Common Regression Evaluation Metrics

To evaluate the performance of a regression model, several metrics can be used, each providing a different perspective on how well the model predicts the target variable. Here are some of the most common ones:

1.  **Mean Absolute Error (MAE)**:
    *   **Definition:** The average of the absolute differences between the predicted values and the actual values.
    *   **Interpretation:** Represents the average magnitude of the errors in a set of predictions, without considering their direction. It's robust to outliers because it uses absolute values.
    *   **Units:** The same units as the dependent variable (e.g., dollars for charges).

2.  **Mean Squared Error (MSE)**:
    *   **Definition:** The average of the squared differences between the predicted and actual values.
    *   **Interpretation:** Penalizes larger errors more heavily than MAE due to squaring. This makes it sensitive to outliers. It is often used because it is differentiable, which is useful for optimization.
    *   **Units:** The square of the units of the dependent variable (e.g., dollars squared).

3.  **Root Mean Squared Error (RMSE)**:
    *   **Definition:** The square root of the MSE.
    *   **Interpretation:** Addresses the unit issue of MSE by bringing the error magnitude back to the same units as the dependent variable, making it more interpretable. It still penalizes large errors more than MAE.
    *   **Units:** The same units as the dependent variable (e.g., dollars).

4.  **R-squared (Coefficient of Determination)**:
    *   **Definition:** Represents the proportion of the variance in the dependent variable that is predictable from the independent variables.
    *   **Interpretation:** Ranges from 0 to 1 (or sometimes negative if the model is worse than a simple mean). An \( R^2 \) of 1 indicates that the model explains all the variability of the dependent variable, while an \( R^2 \) of 0 indicates that the model explains no variability. It's a measure of goodness-of-fit.
    *   **Units:** Unitless.

5.  **Adjusted R-squared**:
    *   **Definition:** A modified version of \( R^2 \) that has been adjusted for the number of predictors in the model.
    *   **Interpretation:** Useful for comparing models with different numbers of predictors. It increases only if the new term improves the model more than would be expected by chance and decreases if a predictor improves the model by less than expected by chance.
    *   **Units:** Unitless.

These metrics will be used to evaluate and compare the performance of the LinearRegression and SGDRegressor models.
